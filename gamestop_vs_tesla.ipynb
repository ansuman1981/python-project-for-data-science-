# **Extract Revenue data through Web Scraping**
now we move to the second most major part of the project:

**Extracting company revenue/profit data from webpages.**

## What is Revenue data?
the money earned by the business company  from its business operations.

## Why compare Revenue with stock data?
Because sometimes:    
* stoc price rises but comapany profit and loss and revenus do not.

**This helps Analysts understand:**
* hype
* overvaluation
* investor behavior

```python
# create revenue dataset
tesla_revenue = pd.DataFrame({
    "Date": ["2024", "2023", "2022", "2021", "2020"],
    "Revenue": ["96,773", "81,462", "53,823", "31,536", "21,461"]
})
```

```python
# view the dataset
tesla_revenue.head()
```

```python
# check dataset info
""" check data types and missing values """
tesla_revenue.info()
```

```python
# clean revenue column
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].str.replace(",","")
```

```python
# convert Revenue to numeric
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].astype(int)
```

```python
# verify the info
tesla_revenue.info()
```

```python
# statistical informations
tesla_revenue.describe()
```
